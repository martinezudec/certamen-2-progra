```python
#  CÓDIGO DEL PROBLEMA 1
import numpy as np

TLDs = []  # Lista para almacenar los TLDs de los correos electrónicos

def read_validate(m, txt): # Lee y valida un valor numérico ingresado por el usuario.
    while True:
        val = input(txt)
        if not val.isnumeric():
            print("Error, el valor debe ser un número")
            continue
        val = int(val)

        if val < m:
            print(f'Error, la cantidad debe ser mayor que {m}')
            continue

        break

    return val

n = read_validate(0, "Ingrese la cantidad de correos a recibir: ")  # Lectura y validación de la cantidad de correos

def find_TLD(correo): # Obtiene el TLD (dominio de nivel superior) de un correo electrónico.
    correo = correo.split(".")
    tld = correo[-1]  # Último elemento de la lista obtenida al dividir el correo por el caracter "."
    return tld

for correos in range(n):
    a = input("Ingrese correo electrónico: ")  # Lectura del correo electrónico
    TLDs.append(find_TLD(a))  # Agregar el TLD del correo electrónico a la lista

tld, freq = np.unique(TLDs[:], return_counts=True)  # Obtener los TLDs únicos y sus frecuencias

# Imprimir el TLD más frecuente y su frecuencia correspondiente
print(f"{tld[np.argmax(freq)]} - {freq[np.argmax(freq)]}")

```

```python
#  CÓDIGO DEL PROBLEMA 2
def validate_roman_number(roman_number):
    # Reglas para validar un número romano
    if 'IIII' in roman_number or 'VV' in roman_number or 'XXXX' in roman_number or 'LL' in roman_number or 'CCCC' in roman_number or 'DD' in roman_number or 'MMMM' in roman_number:
        return False

    if 'IVV' in roman_number or 'IXX' in roman_number or 'IXC' in roman_number or 'XLX' in roman_number or 'XCX' in roman_number or 'XCD' in roman_number or 'CDC' in roman_number or 'CMC' in roman_number:
        return False

    if 'IIV' in roman_number or 'XXL' in roman_number or 'CCD' in roman_number or 'LLD' in roman_number or 'MMD' in roman_number or 'DM' in roman_number or 'LM' in roman_number or 'DM' in roman_number or 'IM' in roman_number:
        return False

    return True


# Lectura de la entrada
roman_number = input("Ingrese un número romano: ")

# Validación del número romano
if validate_roman_number(roman_number):
    print("Número romano bien ingresado")
else:
    print("Número romano mal ingresado")
```

```python
import numpy as np

def generate_alternative_matrix(H):
    # Genera una matriz alternativa a partir de una matriz hueca

    rows, cols = H.shape

    alternative_matrix = []
    count = 0

    for i in range(rows):
        for j in range(cols):
            if H[i][j] != 0:
                alternative_matrix.append([i, j, H[i][j]])
                count += 1

    return np.array(alternative_matrix)


def validate_input(txt, m, M):
    # Valida y solicita una entrada numérica dentro de un rango especificado

    valid_input = False
    while not valid_input:
        input_value = input(txt)
        if not input_value.lstrip('-').isdigit():
            print("Error: Debe ingresar un número entero. Intente nuevamente.")
        else:
            input_value = int(input_value)
            if input_value < m or input_value > M:
                print(f"Error: El valor debe estar en el rango [{m}, {M}]. Intente nuevamente.")
            else:
                valid_input = True

    return input_value


# Lectura de la entrada
f = validate_input("Ingrese el número de filas (entre 5 y 100): ", 5, 100)
c = validate_input("Ingrese el número de columnas (entre 5 y 100): ", 5, 100)

H = np.empty((f, c), dtype=int) # Creación de una matriz vacía con tamaño (f, c)

for i in range(f):
    valid_input = False
    while not valid_input:
        row_input = input(f"Ingrese {c} elementos para la fila {i + 1}, separados por espacios: ") # Se solicita al usuario que ingrese los elementos de la fila i + 1
        row = row_input.split() # Se separa la entrada en una lista de elementos

        if len(row) != c: # Si la cantidad de elementos ingresados no coincide con c
            print(f"Error: Debe ingresar exactamente {c} elementos. Intente nuevamente.") # Se muestra un mensaje de error
        else:
            valid_input = all(element.lstrip('-').isdigit() for element in row) # Se verifica que todos los elementos sean dígitos
            if not valid_input: # Si hay algún elemento no numérico
                print("Error: Los elementos deben ser números enteros. Intente nuevamente.") # Se muestra un mensaje de error

    H[i] = [int(element) for element in row] # Se convierten los elementos de la fila a enteros y se asignan a H[i]

# Generación de la matriz alternativa
A = generate_alternative_matrix(H)

# Impresión de la matriz alternativa
print("Matriz alternativa:")
for row in A:
    print(*row)
```
